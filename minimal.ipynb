# Нахождение минимума функции -cos(x) на отрезке

```python
import math
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
def f(x):
    return -math.cos(x)
```

Построю график функции, чтобы выяснить необходимый отрезок.

```python
x0 = np.linspace(0, 3*math.pi, 200)
y0 = [f(k) for k in x0]
plt.plot(x0, y0)
```

Как видно из построения, функция имеет несколько минимумов.
Выберу отрезок от 4 до 7.

```python
a = 4
b = 7
psi = 1e-10 #точность

pzs = (1+math.sqrt(5))/2 #пропорция золотого сечения
```

А вот и сам метод:

```python
def golden(a, b):
    
    while abs(b-a) >= psi:
        
        x1 = b - (b-a)/pzs
        x2 = a + (b-a)/pzs
        
        y1 = f(x1)
        y2 = f(x2)
        
        if y1 >= y2:
            a = x1
        else: 
            b = x2
    
    x = (a+b)/2
    
    return x
```

```python
print ('Минимум на этом отрезке в точке')
print ('x =', golden(a, b))
print('y =', f(golden(a,b)))
```

```python
xx = np.linspace(a-1, b+1, 200)
yy = [f(i) for i in xx]

xmin = golden(a, b)
ymin = f(golden(a,b))

plt.plot(xx, yy,'r', label = '-cos(x)')
plt.plot(xmin, ymin, 'o', label = 'минимум')
plt.legend()
```
