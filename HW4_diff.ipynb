```python
#plotting and magic 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


def f(x, h):
    return (x+h)**11

"""Считаем производные разными способами"""

def forw_diff_f(x, h): #односторонняя двухточечная система
    return (f(x, h) - f(x, 0))/h

def centr_diff_f(x, h): #центральная двухточечная система
    return ((f(x , h/2) - f(x , - h/2))/h)
  
def diff_f(x): #аналитически
    return 11*x**10
```

```python
#машинный эпселон
eps = 1.0
while 1 + eps > 1:
    eps = eps/2
print (eps)
```

Машинный эпселон примерно равен 10^-16, что пригодится для задания h. h должна быть много меньше 1, поэтому нужно задавать ее в диапазоне от 10^-16 до 1. Для интереса возьму немного меньше.

```python
x = 1.2 #фиксирую х
hi = [ 10 ** (k/10) for k in range ( -200, 20 ) ] 
```

```python
yf = [abs( forw_diff_f(x, i) - diff_f(x)) for i in hi] #central
yc = [abs( centr_diff_f(x, i) - diff_f(x)) for i in hi] #forward

plt.loglog(hi, yc, color='magenta', linestyle='--', marker='o', label = u'Центральная')
plt.loglog(hi, yf, color='turquoise', linestyle='--',  marker='o', label = u'Односторонняя')

#Style ヾ(⌐■_■)ノ♪
plt.ylabel('модуль разницы производных')
plt.xlabel('h')
plt.title('Разница в вычислении производной')
plt.grid(True)
plt.legend()
plt.show()
```

Отсюда можно заметить, что при h < 10^-16 значения становятся равны константе (10^2). Это происходит потому, что обнуляется производная с h меньше, чем машинный эпсилон. Далее виден большой разброс значений - неточности, получаемые при вычислении слишком маленьких h и округлении. В конце мы можем заметить тенденцию: чем дальше, от h ≈ 10^-7, тем сильнее погрешность вычиcлени. Также из графика очевидно превосходство вычисления производной с помощью центральной двухточечной системы над вычислением с помощью односторонней, так как минимум функции оказался меньше. От h ≈ 0 оба способа дают примерно одинаковый плохой результат.
